```python
import json
import seaborn as sns
import pandas as pd
from matplotlib import pyplot
```

```python
file = open('../benchmark/finalReport.json')
data = json.load(file)

gatsby_data = []
next_data = []
react_data = []

for run in data:
    gatsby_data.append(run['gatsby'])
    next_data.append(run['next'])
    react_data.append(run['react'])

df_gatsby = pd.DataFrame(gatsby_data)
df_next = pd.DataFrame(next_data)
df_react = pd.DataFrame(react_data)
```

```python
# Some options that should apply to every plot we make
sns.set_theme(style="darkgrid")
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python

lcp_df = pd.DataFrame({
    'gatsby': df_gatsby['largestContentfulPaint'],
    'next': df_next['largestContentfulPaint'],
    'react': df_react['largestContentfulPaint'],
})


ax = sns.lineplot(data=lcp_df)
ax.set(xlabel='runs', ylabel='milliseconds')
ax.set_title('Largest contentful paint')
```

```python
tbt_df = pd.DataFrame({
    'gatsby': df_gatsby['totalBlockingTime'],
    'next': df_next['totalBlockingTime'],
    'react': df_react['totalBlockingTime'],
})


ax = sns.lineplot(data=tbt_df)
ax.set(xlabel='runs', ylabel='milliseconds')
ax.set_title('Total blocking time')
```
